# Analiza Telco Customer Churn podataka

Ovaj notebook prikazuje osnovnu analizu podataka i pripremu modela za predikciju churn-a korisnika.

## 1. Učitavanje biblioteka

```python
import pandas as pd
import matplotlib.pyplot as plt

from izvorni_kod.ucitavanje_podataka import ucitaj_podatke
from izvorni_kod.priprema_podataka import pripremi_podatke_za_model
from izvorni_kod.predikcija import napravi_predikcije_za_sve_korisnike
from izvorni_kod.objasnjenja import dodaj_objasnjenja_i_akcije
from izvorni_kod.roi import dodaj_roi_kolone, napravi_roi_summary
```

## 2. Učitavanje podataka

```python
df, info = ucitaj_podatke()
print(info)
df.head()
```

## 3. Osnovne informacije o datasetu

```python
print('Broj korisnika:', len(df))
print('Broj kolona:', len(df.columns))
print('Kolone:')
print(df.columns.tolist())
```

## 4. Churn distribucija

```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts().plot(kind='bar')
plt.title('Distribucija churn korisnika')
plt.xlabel('Churn')
plt.ylabel('Broj korisnika')
plt.show()
```

## 5. Priprema podataka za model

```python
podaci = pripremi_podatke_za_model(df)
print('X_train:', podaci['X_train'].shape)
print('X_test:', podaci['X_test'].shape)
print('Numeričke kolone:', podaci['numericke_kolone'])
print('Kategorijske kolone:', podaci['kategorijske_kolone'])
```

## 6. Predikcije churn rizika

```python
df_predikcije = napravi_predikcije_za_sve_korisnike(df)
df_predikcije[['customerID', 'churn_percent', 'risk_level', 'expected_loss']].head(10)
```

## 7. Objašnjenja i preporučene akcije

```python
df_predikcije = dodaj_objasnjenja_i_akcije(df_predikcije)
df_predikcije[['customerID', 'churn_percent', 'reasoning', 'recommended_action']].head(10)
```

## 8. ROI analiza

```python
df_predikcije = dodaj_roi_kolone(df_predikcije)
summary = napravi_roi_summary(df_predikcije.head(100))
summary
```

## 9. Zaključak

Analiza pokazuje da se na osnovu podataka o korisnicima može procijeniti rizik odlaska iz kompanije. Model omogućava rangiranje korisnika po riziku, objašnjenje glavnih faktora i predlaganje retention akcija.
